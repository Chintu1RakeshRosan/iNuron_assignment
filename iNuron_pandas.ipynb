# 1st

```python
import pandas as pd
import numpy as np
df = pd.DataFrame({
    'From_To': ['LoNDon_paris', 'MAdrid_miLAN','londON_StockhOlm','Budapest_PaRis', 'Brussels_londOn'],
    'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],
    'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]],
    'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )','12. Air France', '"Swiss Air"']})
```

```python
df
```

```python
df['PreviousFlightNumber']=df.FlightNumber.shift(1)
```

```python
df['FlightNumber'] = df['FlightNumber'].fillna(df['PreviousFlightNumber'] + 10)
```

```python
df
```

```python
df.drop('PreviousFlightNumber',axis='columns',inplace=True)
```

```python
df
```

```python
df['FlightNumber'] = df['FlightNumber'].fillna(10)
```

```python
df
```

```python
df.dtypes
```

```python
df['FlightNumber'] = df['FlightNumber'].apply(int)
```
